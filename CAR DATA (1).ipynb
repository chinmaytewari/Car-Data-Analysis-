```python
import numpy as np 
import pandas as pd 
import seaborn as sns
from matplotlib import pyplot as plt


car = pd.read_csv(r"C:\Users\chinmay\Desktop\git in one folder\cars data.csv")
car.head()
```

```python
car.shape
```

```python
car.drop(['MPG_Highway','MPG_City','EngineSize','Horsepower','Weight','Wheelbase','Length'],axis=1,inplace=True)
car.tail(5)
```

```python
car.info()
```

```python
car.rename(columns = {'Invoice':'Selling Price'}, inplace = True)
car
```

```python
d = car["Cylinders"].value_counts()
d
```

```python
a = car["DriveTrain"].value_counts()
a
```

```python
x = dict(a)


drive = list(x.keys())
values = list(x.values())
  
fig = plt.figure(figsize = (80, 20))
 
# creating the bar plot
plt.bar(drive, values, color ='blue',
        width = 0.5)
 
plt.xlabel("Type ",size=100)
plt.ylabel("Number of Models",size =100)
plt.title("Drive Train",size = 100)
plt.xticks(size = 70)
plt.yticks(size = 70)

plt.show()
```

```python
z = car["Origin"].value_counts()
z
```

```python
t = dict(z)
# t
m = list(t.keys())
n  = list(t.values())
i=0
while i in range(0,3):
    plt.bar(m[i],n[i])
    i=i+1
plt.xlabel("Origin ",size=20)
plt.ylabel("Number of Vehicles",size =20)

plt.show()
```

```python
b = car["Make"].value_counts()
b
```

```python
b.unique()
```

```python
w = dict(b)
w
```

```python
company = list(w.keys())
number_of_models = list(w.values())
fig = plt.figure(figsize = (80, 20))

i=0
for i in range(0,18):
    plt.bar(company[i], number_of_models[i],
        width = 0.5)
 
plt.xlabel("company name",size=70)
plt.ylabel("Number of Models",size = 70)

plt.xticks(size = 50,rotation = 90)
plt.yticks(size = 50)

plt.show()
```

```python
x = car["Type"].value_counts()
# x
a = x.values

l = ['Sedan', 'SUV' ,'Sports','Wagon' ,'Truck' ,'Hybrid']
plt.pie(a,labels=l,autopct='%1.2f%%')
plt.title("Market share of different Type of Vehicle")
plt.show()
```
